```python
#plotting using matplotlib
plt.scatter(x,y,c='g')
#c basically means the colour
#ussed to scatter values in a 2d plot
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
#we need not write plt.show()
```

```python
y = x*x
plt.plot(x,y,'g*--',linestyle='dashed',linewidth=2,markersize=8)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
#creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
```

```python
x= np.arange(1,11)
y  = 3*x+5
plt.plot(x,y,'r*--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot')
```

```python
np.pi
```

```python
x = np.arange(0,4*np.pi,0.1)
#0.1 is the step size
y = np.sin(x)
plt.plot(x,y)
```

```python
x= np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine')
plt.subplot(2,1,1)
plt.plot(x,y_cos)
plt.title('Cosine')
```

```python
x = [2,8,10]
y = [11,16,9]
x2 = [3,9,11]
y2=[6,15,7]
plt.bar(x,y,width=0.5,align='center')
plt.bar(x2,y2,color='g')
plt.title('barGraph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
#Histogram used to find densities basically
#Box plot helps to find percentiles
data = [np.random.normal(0,std,100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True,)
```

```python
#piecharts
labels = 'Python','c++','Ruby','Java'
sizes=[215,130,245,210]
colors=['Gold','Yellowgreen','lightcoral','lightskyblue']
explode=(0.3,0.1,0.5,0.2)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,shadow=True,autopct='%1.1f%%')
plt.axis('equal')
```
